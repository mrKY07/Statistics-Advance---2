# THEROTICAL QUESTIONS


1.What is hypothesis testing in statistics?

Hypothesis testing is a method in statistics to check if a claim or guess about a population is true using data.
We start with a basic assumption (called a hypothesis), then collect data and see if the data supports that assumption or not.
It helps in making decisions using sample data.

For example

 if I want to check if a coin is fair, I can use hypothesis testing after tossing it many times.



2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

The null hypothesis (H0) is the default or starting assumption, like "nothing has changed" or "there is no effect."
The alternative hypothesis (H1) is what we want to prove, like "something has changed" or "there is an effect."
We test the data to see if we can reject the null and accept the alternative.

For example

H0: A medicine doesn’t work. H1: The medicine works.



3.What is the significance level in hypothesis testing, and why is it important?

The significance level (written as α) is the chance we are okay with being wrong when we reject the null hypothesis.
It’s usually set at 0.05 or 5%, meaning we accept a 5% chance of making a mistake.
It helps us decide how strong the evidence must be to reject the null.
If the chance of error is too high, we don’t reject the null.
It’s important to control the risk of making wrong decisions.



4.What does a P-value represent in hypothesis testing?

The P-value tells us the probability of getting results as extreme as our data, if the null hypothesis is true.
A small P-value means the result is unlikely under the null, so we may reject it.
It helps us decide if the data is strong enough to support the alternative hypothesis.

For example

 a P-value of 0.01 means there’s only a 1% chance of getting such results if H0 is true.



5.How do you interpret the P-value in hypothesis testing?

If the P-value is less than the significance level (α), we reject the null hypothesis.
If the P-value is greater than α, we fail to reject the null hypothesis (we keep H0).

For example, if α = 0.05 and P = 0.02, then we reject H0 because 0.02 < 0.05.
It means the result is statistically significant.
So, a smaller P-value gives stronger evidence against the null.

6.What are Type 1 and Type 2 errors in hypothesis testing?

A Type 1 error happens when we reject the null hypothesis even though it is actually true.
A Type 2 error happens when we don’t reject the null even though it is actually false.
Type 1 is like a false alarm, and Type 2 is like missing a real issue.

Example:

 Saying a medicine works (Type 1) when it doesn’t, or saying it doesn’t work (Type 2) when it does.



7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

A one-tailed test checks for an effect in one direction only (greater than or less than).
A two-tailed test checks in both directions (not equal to).
Use one-tailed if you only care about one side, and two-tailed if any change matters.

Example:

Testing if a new battery lasts longer (one-tailed) or just different (two-tailed).



8.What is the Z-test, and when is it used in hypothesis testing?

A Z-test is used to check if there’s a significant difference between sample and population.
We use it when the sample size is large (usually n > 30) and the population standard deviation is known.
It helps to test means or proportions.

Example:

 Checking if average marks in a class are equal to the national average.



9.How do you calculate the Z-score, and what does it represent in hypothesis testing?

The formula is:
Z = (X - μ) / σ
Where X is the sample mean, μ is the population mean, and σ is the standard deviation.
It shows how many standard deviations the sample is from the mean.
A high Z-score means the sample is far from the mean, which might lead to rejecting the null.



10.What is the T-distribution, and when should it be used instead of the normal distribution?

The T-distribution is used when the sample size is small (n < 30) and the population standard deviation is unknown.
It looks like the normal distribution but has fatter tails.
This helps with more uncertainty in small samples.
So, we use t-test instead of z-test when we don’t know σ and have small data.

11.What is the difference between a Z-test and a T-test?

The Z-test is used when the sample size is large and the population standard deviation is known.
The T-test is used when the sample size is small and the standard deviation is unknown.
Z-test uses the normal distribution, while T-test uses the t-distribution.
T-test is better for small and more uncertain data.



12.What is the T-test, and how is it used in hypothesis testing?

The T-test is used to compare sample data to a known value or another sample.
It helps check if the difference between means is real or just by chance.
We use it when the sample is small and we don’t know the population standard deviation.

Example:

Testing if the average height of 10 students is different from a known average.



13.What is the relationship between Z-test and T-test in hypothesis testing?

Both Z-test and T-test are used to test hypotheses about means.
They are very similar but used in different conditions.
If the sample is large and σ is known, use Z-test.
If the sample is small and σ is unknown, use T-test.
As the sample size grows, T-distribution becomes similar to normal distribution, so T and Z become close.



14.What is a confidence interval, and how is it used to interpret statistical results?

A confidence interval (CI) is a range that likely contains the true population value.
It is written like: “We are 95% confident the true mean is between A and B.”
It helps us understand the accuracy of our estimate.
If a confidence interval includes the value from H0, we usually don’t reject H0.



15.What is the margin of error, and how does it affect the confidence interval?

The margin of error shows how much the sample estimate might be off from the true value.
It depends on sample size and confidence level.
Larger margin = wider interval = less precise.
Smaller margin = narrower interval = more precise.
So, margin of error controls the width of the confidence interval.

16.How is Bayes' Theorem used in statistics, and what is its significance?

Bayes' Theorem is used to update the probability of an event based on new information.
It helps us make better decisions when we have prior knowledge.

Formula:

P(A|B) = [P(B|A) × P(A)] / P(B)

It’s useful in areas like medical testing, spam filters, and machine learning.
Example: If a test is positive, Bayes’ Theorem helps find the chance that a person actually has the disease.





17.What is the Chi-square distribution, and when is it used?

The Chi-square distribution is used to test relationships between categorical variables.
It is not symmetric and always positive.
It is used in tests like the Chi-square test of independence and goodness of fit.

Example:

Checking if people’s choice of pet depends on their city.



18.What is the Chi-square goodness of fit test, and how is it applied?

The Chi-square goodness of fit test checks if observed data fits a known or expected distribution.
We compare observed counts with expected counts and calculate a Chi-square value.
If the value is big, we reject the null hypothesis.

Example:

Tossing a die 60 times to check if it’s fair (expect 10 of each number).



19.What is the F-distribution, and when is it used in hypothesis testing?

The F-distribution is used to compare two variances or in ANOVA tests.
It is right-skewed and only positive.
If the F-value is large, it means the groups are very different.
Used when we want to check if different samples have equal variances.



20.What is an ANOVA test, and what are its assumptions?

ANOVA (Analysis of Variance) checks if three or more group means are significantly different.
It uses the F-test to compare group variances.

Assumptions:
Samples are independent
Data is normally distributed
Variances are equal

Example:

 Checking if three teaching methods give different results.



21.What are the different types of ANOVA tests?

There are mainly three types of ANOVA:

One-way ANOVA – compares means of 1 factor with 3+ groups.

Two-way ANOVA – compares means with 2 factors.
Repeated Measures ANOVA – used when the same subjects are tested more than once.
Each is used based on how many factors and groups we are comparing.



22.What is the F-test, and how does it relate to hypothesis testing?

The F-test compares two variances to see if they are significantly different.
It is used in ANOVA and in checking model accuracy.
If the F-value is large, we may reject the null hypothesis.
It helps decide if group differences are real or just by chance.

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = np.array([52, 55, 48, 50, 53, 54, 49, 51, 47, 56])
sample_mean = np.mean(sample)
sample_size = len(sample)
population_mean = 50  # Null hypothesis mean
population_std = 5    # Known population std dev

# Calculate Z-statistic
z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z-statistic:", z)
print("P-value:", p_value)

# Decide to reject or not reject the null hypothesis at 5% significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

sample = [102, 99, 100, 105, 98, 101, 103]
pop_mean = 100
pop_std = 5
n = len(sample)
sample_mean = np.mean(sample)

z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label='Standard Normal Distribution')
plt.fill_between(x, 0, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.axvline(-z_score, color='blue', linestyle='--')
plt.title('Two-tailed Z-test')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.3f}")

if p_value < alpha:
    print("Reject null hypothesis: Sample mean is significantly different.")
else:
    print("Fail to reject null hypothesis: No significant difference.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type_errors(mu0=0, mu1=1, sigma=1, alpha=0.05):
    z_alpha = norm.ppf(1 - alpha)
    x = np.linspace(-4, 6, 1000)

    h0 = norm.pdf(x, mu0, sigma)
    h1 = norm.pdf(x, mu1, sigma)

    plt.plot(x, h0, label='H0 Distribution (mean=0)')
    plt.plot(x, h1, label='H1 Distribution (mean=1)')

    plt.fill_between(x, 0, h0, where=(x > z_alpha), color='red', alpha=0.3, label='Type 1 Error (α)')
    plt.fill_between(x, 0, h1, where=(x <= z_alpha), color='blue', alpha=0.3, label='Type 2 Error (β)')

    plt.axvline(z_alpha, color='black', linestyle='--', label=f'Critical value (z={z_alpha:.2f})')
    plt.title('Type 1 and Type 2 Errors')
    plt.xlabel('Test Statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()

plot_type_errors()
```

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_crit * std_err
    lower = mean - margin
    upper = mean + margin
    return lower, upper

sample = [12, 15, 14, 10, 13, 12, 16, 11, 14]
ci_lower, ci_upper = confidence_interval(sample)
print(f"Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

```

```python
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    numerator = likelihood_B_given_A * prior_A
    denominator = numerator + likelihood_B_given_not_A * prior_not_A
    posterior_A_given_B = numerator / denominator
    return posterior_A_given_B

prior_A = 0.01
likelihood_B_given_A = 0.9
prior_not_A = 0.99
likelihood_B_given_not_A = 0.05

posterior = bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A)
print(f"Posterior Probability P(A|B): {posterior:.4f}")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

data = {'Gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Female', 'Male'],
        'Preference': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B']}

df = pd.DataFrame(data)
contingency_table = pd.crosstab(df['Gender'], df['Preference'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2:.3f}")
print(f"P-value: {p:.3f}")

alpha = 0.05
if p < alpha:
    print("Reject null hypothesis: Variables are dependent.")
else:
    print("Fail to reject null hypothesis: Variables are independent.")
```

```python
import numpy as np
import pandas as pd

observed = np.array([[30, 10],
                     [20, 40]])

row_sums = observed.sum(axis=1, keepdims=True)
col_sums = observed.sum(axis=0, keepdims=True)
total = observed.sum()

expected = row_sums @ col_sums / total
expected_df = pd.DataFrame(expected, columns=['Category1', 'Category2'], index=['Group1', 'Group2'])

print("Expected Frequencies:")
print(expected_df)
```

```python
from scipy.stats import chisquare

observed = [50, 30, 20]
expected = [40, 40, 20]

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: Observed data does not fit expected distribution.")
else:
    print("Fail to reject null hypothesis: Observed data fits expected distribution.")
```
